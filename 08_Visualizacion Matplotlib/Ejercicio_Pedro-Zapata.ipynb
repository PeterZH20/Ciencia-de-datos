# Ejercicio 3: Pacientes confirmados con Covid en México


Fuente: https://datos.covid-19.conacyt.mx/#DownZCSV

```python
df= pd.read_csv("datos/Casos_Diarios_Estado_Nacional_Confirmados_20201129.csv")
df=df.loc[:,'nombre':]
df = df.set_index('nombre').T
df=df.reset_index()
df = df.rename(columns = {'index': 'Fecha'})
df['Fecha'] = pd.to_datetime(df['Fecha'], format= '%d-%m-%Y')
df = df.set_index('Fecha')
df.head()
```

```python
df.columns
```

<div class="alert alert-success">
    
1. Hacer una grafica de barras comparando todos los meses, considerando el total nacional (por mes) 
    
</div> 

```python
df.groupby(df.index.month)['Nacional'].sum()
```

```python
#Estilo orientado a objetos
fig, ax = plt.subplots(figsize=(6,3), dpi=200)

ax.bar(df.groupby(df.index.month)['Nacional'].sum().index, 
       df.groupby(df.index.month)['Nacional'].sum().values,
      color='green')          
ax.set(title='Total nacional por meses', xlabel= 'Mes' , ylabel='Total de contagios')

plt.show()
```

```python
#Estilo Matlab
plt.figure(figsize=(6,3), dpi=200)

plt.bar(df.groupby(df.index.month)['Nacional'].sum().index, 
       df.groupby(df.index.month)['Nacional'].sum().values,
      color='red')  
plt.title('Total nacional por meses',fontsize=10)
plt.xlabel('Mes',fontsize=10)
plt.ylabel('Total de contagios',fontsize=10);

plt.show()
```

<div class="alert alert-success">
    
2. Hacer una grafica de barras comparando  los meses de agosto, septiembre, octubre y noviembre, considerando solo las entidades de Yucatán, Distrito Federal  y Jalisco
    
</div> 

```python
#df.groupby(df.index.month)[['YUCATAN','DISTRITO FEDERAL','JALISCO']].sum().tail(4)['YUCATAN']
```

```python
labels=['Ago','Sep','Oct','Nov']
ancho=0.5
p=[]
estados=['YUCATAN','DISTRITO FEDERAL','JALISCO']
for estado in estados:
    d=[]
    i=0
    for i in range(4):
        d.append(df.groupby(df.index.month)[['YUCATAN','DISTRITO FEDERAL','JALISCO']].sum().tail(4)[estado].values[i])
        i+=1
    p.append(d)
p
```

```python
co=np.arange(len(labels))
#Estilo Orientado a Objetos
fig, ax = plt.subplots(figsize=(8,5))

ax.bar(co - 0.33*ancho, p[0], color = 'r', width = ancho/4,  label='Yucatán')
ax.bar(co             , p[1], color = 'b', width = ancho/4,  label='Distrito Fed.')
ax.bar(co + 0.33*ancho, p[2], color = 'g', width = ancho/4,  label='Jalisco')



ax.set_xticks(co)
ax.set_xticklabels(labels)
ax.legend()
ax.set(xlabel= 'Meses' , ylabel='Total de contagios')

plt.show()
```

```python
#Estilo Matlab
plt.figure(figsize=(8,5))  

plt.bar(co - 0.33*ancho, p[0], color = 'r', width = ancho/4,  label='Yucatán')
plt.bar(co             , p[1], color = 'b', width = ancho/4,  label='Distrito Fed.')
plt.bar(co + 0.33*ancho, p[2], color = 'g', width = ancho/4,  label='Jalisco')

plt.xticks( co , labels )
plt.xlabel('Meses')
plt.ylabel('Total de contagios')
plt.legend()

plt.show()
```
